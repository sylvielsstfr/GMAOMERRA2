```python
import os,sys
import pandas as pd
import numpy as np
```

```python
path='datasets_2017'
```

```python
files=['MERRA2_2017_inst1_2d_asm_Nx_M2I1NXASM_ctio_AllYear.csv',
       'MERRA2_2017_tavg1_2d_aer_Nx_M2T1NXAER_ctio_AllYear.csv',
       'MERRA2_2017_tavg1_2d_rad_Nx_M2T1NXRAD_ctio_AllYear.csv']
```

```python
output_file='MERRA2_2017_M2I1NXASM_M2T1NXAER_M2T1NXRAD_ctio_AllYear.csv'
```

# Read independant dataset

```python
df1=pd.read_csv(os.path.join(path,files[0]))
df2=pd.read_csv(os.path.join(path,files[1]))
df3=pd.read_csv(os.path.join(path,files[2]))
```

# Specify which column is the index

- the index is the time

```python
df1.set_index("time",inplace=True)
df1.head()
```

```python
df2.set_index("time",inplace=True)
df2.head()
```

```python
df3.set_index("time",inplace=True)
df3.head()
```

# concatenate in vertical the 2 first dataset and reorder by index

```python
df12=pd.concat([df1,df2]).sort_index()
df12
```

```python
df123=pd.concat([df12,df3], axis=1).sort_index()
```

```python
#df=df123.reindex_axis(['ps','pwv','ozone',
#                    'TOTEXTTAU','TOTANGSTR','TOTSCATAU',
#                    'TAUTOT','TAUHGH','TAUMID','TAULOW'], axis=1)
df=df123.reindex(['ps','pwv','ozone',
                  'TOTEXTTAU','TOTANGSTR','TOTSCATAU',
                  'TAUTOT','TAUHGH','TAUMID','TAULOW'], axis=1)
df
```

```python
df.to_csv("")
```
